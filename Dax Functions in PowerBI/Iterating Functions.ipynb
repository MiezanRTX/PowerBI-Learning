## Practice with Iterating Functions

Iterating functions are great for performing computations row by row and should be used when you don't want to aggregate all the values at once for computation. In this exercise, you are trying to find out the profit ratio for a given employee. **The profit ratio is a performance indicator in business analytics that tells how well someone or something is performing**. For the profit ratio we need to perform a row-wise analysis to divide the total profits by the total revenue.

## More iterating functions

Your manager is pleased by your current employee analysis, and asks you to dive deeper by looking at the **average discount rate** given per order by employees. Since the Retail Price and selling price (i.e. Unit Price) are known, you can **calculate the given discount by subtracting the unit price by the retail price and dividing it by the retail price**. This expression is perfect to use in an iterating function, so that you don't need to create an intermediate measure first.

### Steps

1. Create a new measure, called Average Discount Rate. Use AVERAGEX() on the Fact_Sales table while subtracting the Unit Price from Retail Price, and dividing the difference by Retail Price. Format as percentage.

2. Create a Scatter chart visual, with Average Discount Rate on the X Axis, and Profit Ratio on the Y Axis. Color by Dim_Employee[Employee].
