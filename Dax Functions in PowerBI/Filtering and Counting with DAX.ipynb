# Calculating with a filter
When performing analysis for business decisions it's important to analyze data from different perspectives to find trends/issues. In this example, World Wide Importers **recently introduced items that required refrigeration** (i.e. chilled items). You're asked to look into this further **to see if any trends exist among salespersons in the sale of these items**. You'll do that by looking at what **percent of chiller item sales are part of total sales for a salesperson.**

### Code Explaination
- Goal is to find total sale where product is chiller type 
- However Fact table doesnt have this info but dimension table stock item does.
- Use RELATED function so that DAX look at related dimension table which is Dim_StockItem
    - The **FactSales** is connected with Dim_StockItem is connected with **StockItemID**
- From here the DAX able to look at Dim_StockItem column call [Is Chiller Stock] and we filter where is TRUE.
