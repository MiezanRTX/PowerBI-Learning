# Controlling the import process

You have been informed by the WWI IT department that the old MySQL server is being deprecated soon and you will need to access the data through another database management system - Open DataBase Connectivity (ODBC).

This is an open standard Application Programming Interface (API) that allows you to create and store different database connections within one interface. It therefore also has different characteristics when being used to connect with Power BI. Let's see how we can use this new connection type in the report we are building.
