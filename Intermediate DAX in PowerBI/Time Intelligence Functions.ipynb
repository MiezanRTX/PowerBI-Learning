# Time Intelligence Functions

## Use of TOTALYTD()

### Context
One of the main uses of time intelligence functions is that they allow you to calculate measures that compare values over periods of time. One such calculation is the year-to-date (YTD) calculation, which returns the result of an expression (for example the running sum), from the beginning of the year until today. In this scenario, you will go back in time and use a time intelligence function to calculate and visualize the year to date sales of the World Wide Importers data.

### Steps
1. Create a new page, called TotalYTD Sales.
Mark the Dim_InvoiceDate table as a date table.
        

2. Use the TOTALYTD() function to calculate the Total Sales year-to-date in a new measure called Total Sales YTD, based on your specified date table.
Format it as a currency.
          

3. Create a Line chart visual, with the Total Sales YTD by Month and a line for each Calendar Year. Make sure to use date values from the Dim_InvoiceDate date table.
        



4. What was the year-to-date value for total sales in October 2014?

    $41,545,633
