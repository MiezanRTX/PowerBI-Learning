# SUMMARIZE() the facts

Data tables sometimes contain repetitive rows of information and need to be condensed down for reporting. The SUMMARIZE() function can be used to group a table by one or more columns, condensing tables down. 

In this scenario, you're going to have a look at the Profit Ratio from Chapter 2 again, and **visualize how it changed over time for the California sales**. One approach to do this, is by **creating a new table containing only the data you need**, without the need of specifying an additional relationship, improving performance of your report.

### Steps

1. Go to the Table view and create a new table, called CaliforniaPR (for Profit Ratio's in California only).
This new table should summarize the Fact_Sales table grouping by Employee, Calendar Year, and State Province.You'll filter for "California" later.
    


2. Edit the CaliforniaPR table syntax using a FILTER() function so that it only keeps data from the state "California".
        
    

3. How many rows does your new table have?
        
    


## ADDCOLUMNS()? No problem!

Although it is possible to create new columns with SUMMARIZE(), best practice is to add new columns to a table using the ADDCOLUMNS() function. In this way, the output will be consistent since ADDCOLUMNS() will always evaluate over the row context by default. In this exercise, you'll expand the CaliforniaPR table with the Profit Ratio from Chapter 2.

### Steps

1. Add a column called Profit Ratio to the CaliforniaPR table, using the ADDCOLUMNS() function. You can just reuse the [Profit Ratio] calculated measure, writing all inside this new function.

    

    


2. Create a Ribbon chart visual of the CaliforniaPR data, with a ribbon for every Employee, showing the Profit Ratio by Calendar Year.
