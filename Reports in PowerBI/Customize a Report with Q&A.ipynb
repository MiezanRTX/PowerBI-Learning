# Customize a Report with Q&A

# Add the Q&A visual

Users are enjoying the Wake County Watchdog report so far but we are getting in more and more requests for information. We would like to investigate the idea of allowing report users to engage with Power BI's Q&A visual, starting with the Expenditure Lines report. Depending on how it goes, we might incorporate more Q&A capabilities in other sections of the report.





# Working with the Q&A visual

The Q&A visual takes its best guess at the best visual to use when displaying results. Sometimes, however, we may wish to specify the kind of visual it should display. In this exercise, you will work with different requests for visuals and gain a better understanding of how the Q&A visual can serve user needs.





# Add synonyms for fields

The easiest way to make the Q&A visual usable for your report viewers is to have well-named fields on each table. Even so, however, people may wish to describe a column in multiple ways. This is where synonyms come in. In this exercise, you will add synonyms to three columns, two of which you will use here and one in the next exercise.











# Teach Q&A new terms

Business users often have their own jargon: words and phrases which make sense to them but not to the general public. The Q&A visual will not understand this jargon by default but we can train it to have a better understanding of specific terms. This feature can be rather finicky, especially when dealing with multi-word phrases, but in this exercise, we will see how we can add custom nouns and adjectives to make it easier for business users to define their queries.
