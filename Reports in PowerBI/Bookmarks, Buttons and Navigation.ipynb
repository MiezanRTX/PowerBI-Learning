# Create bookmarks

Bookmarks are a great way of highlighting particular elements of a story, including the page, filter and slicer settings, and active report elements. In this exercise, we would like you to begin creating bookmarks for a review of certain county infrastructure projects, starting with public safety.










# Storytelling with bookmarks

Our watchdog organization wants to ensure that the county is prudently managing finances on capital improvement projects--that is, large-scale purchases of equipment, buildings, and tools. We want to show how easy it is for cost overruns to occur, and we can use bookmarks to tell this cohesive story. Going even further, the bookmark navigator allows us to view bookmarks like a slide deck, moving easily between bookmarks.
