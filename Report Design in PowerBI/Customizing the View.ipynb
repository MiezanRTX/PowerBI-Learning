# Change default font sizes

The Wake County Watchdog team has started going on tour, talking to county agencies about fiscal prudence. Along the way, one user pointed out difficulty distinguishing specific colors on charts. They want you to ensure that people with a color vision deficiency may distinguish colors in our charts. To do so, we will build off of an existing theme called "Colorblind safe".

More general information about report themes can be found in this MSFT Learn article.

# Alter the filter pane

Power BI will allow you to customize settings using both JSON and its customization tool. One of the settings we can **customize is how the filter pane appears**. Management would like to **make the filter pane a little more obvious**, as well as **increasing the font size**.

![Screenshot 2025-12-25 133351.png](<attachment:Screenshot 2025-12-25 133351.png>)

Set the filter pane background color to #A6BEB1 with a transparency of 20%, a title font size of 14pt, and a header font size of 12pt.
