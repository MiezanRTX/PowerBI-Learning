# Choose between visuals

Although there are three charts on the page, we can only see one because it overlaps the other two. Instead of squeezing these charts into a small area, we can use progressive disclosure techniques and a specific implementation involving buttons and bookmarks to show one chart at a time.

In this exercise, we will create the buttons and bookmarks. Unlike the buttons to show and hide departments, we need all three buttons to remain on screen, as a user may wish to navigate between the three options. Therefore, we also need a way of distinguishing which button's chart is visible.
