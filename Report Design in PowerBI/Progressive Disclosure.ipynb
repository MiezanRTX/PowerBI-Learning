# Budgets vs expenditures over time

The Wake County Watchdog team is excited about shifting our Over Budget report from a dashboard-like experience toward progressive disclosure. They have given us one more data request to add to the report. They would like to see the amount spent to date for the fiscal year versus the total budgeted amount, providing us a view of when departments tend to run out of money or go over budget.

Their design staff has started some of the work on progressive disclosure and would like you to continue the job.

# Show and hide departments

One element of progressive disclosure we will incorporate involves hiding information we do not need. In this case, we will use a pair of buttons to perform this operation. One button will show the Departments table, and the other will hide it. To reduce confusion, we will only have one button visible at a time, so if the Department table is visible, the button to hide the table will be visible, but the button to show the table will be hidden.
